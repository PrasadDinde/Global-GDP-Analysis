## GDP Analysis Percentage

```python
df_arab.head()
```

```python
data=df_arab.values
gdp_chanage=[0]
for i in range(1,len(data)):
    curr=data[i][3]
    prev=data[i-1][3]
    gdp_chanage.append(round((curr-prev)/prev*100,2))
df_arab.assign(GDP=gdp_chanage)
```

## GDP Analysis for India

```python
data2=df_ind.values
gdp=[0]
for i in range(1,len(data2)):
    curr=data2[i][3]
    prev=data2[i-1][3]
    gdp.append(round((curr-prev)/prev*100,2))
df_ind.assign(GDP=gdp)
```

## GDP Analysis on whole dataset

```python
final=[]
for country in df['Country Name'].unique():
    df_all=df[df['Country Name']==country]
    data=df_all.values
    gdp=[0]
    for i in range(1,len(data)):
        curr=data[i][3]
        prev=data[i-1][3]
        gdp.append(round((curr-prev)/prev*100,2))
    df_all=df_all.assign(GDP=gdp)
    final.append(df_all)
        
```

```python
df=pd.concat(final,axis=0)
```
